```python
# Average Novelty Scores for Human vs. AI Hypotheses

import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/content/evaluated_hypotheses.csv')

avg_human_novelty = data['human_novelty'].mean()
avg_ai_novelty = data['ai_novelty'].mean()

categories = ['Human', 'AI']
avg_scores = [avg_human_novelty, avg_ai_novelty]

colors = ['#ed6242', '#292273']

plt.figure(figsize=(8, 6))
plt.bar(categories, avg_scores, color=colors)
plt.xlabel('Source', fontsize=12)
plt.ylabel('Average Novelty Score', fontsize=12)
plt.title('Average Novelty Scores for Human vs. AI Hypotheses', fontsize=14)
plt.ylim(0, 10)

for i, score in enumerate(avg_scores):
    plt.text(i, score + 0.2, f'{score:.2f}', ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Average Scores for Human vs AI Hypotheses Across All Categories
human_scores = [
    'human_novelty',
    'human_excitement',
    'human_feasibility',
    'human_effectiveness'
]
ai_scores = [
    'ai_novelty',
    'ai_excitement',
    'ai_feasibility',
    'ai_effectiveness'
]

avg_human_score = data[human_scores].mean().mean()
avg_ai_score = data[ai_scores].mean().mean()

categories = ['Human', 'AI']
avg_scores = [avg_human_score, avg_ai_score]

colors = ['#ed6242', '#292273']

plt.figure(figsize=(8, 6))
plt.bar(categories, avg_scores, color=colors)
plt.xlabel('Source', fontsize=12)
plt.ylabel('Average Score Across Categories', fontsize=12)
plt.title('Average Scores for Human vs AI Hypotheses Across All Categories', fontsize=14)
plt.ylim(0, 10)

for i, score in enumerate(avg_scores):
    plt.text(i, score + 0.2, f'{score:.2f}', ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Summary
human_avg_scores = data[human_scores].mean()
ai_avg_scores = data[ai_scores].mean()

human_overall_avg = human_avg_scores.mean()
ai_overall_avg = ai_avg_scores.mean()

category_names = ['Novelty', 'Excitement', 'Feasibility', 'Effectiveness']
avg_scores_df = pd.DataFrame({
    'Category': category_names,
    'Average Human Score': [f'{score:.2f}' for score in human_avg_scores],
    'Average AI Score': [f'{score:.2f}' for score in ai_avg_scores]
})

overall_row = pd.DataFrame({
    'Category': ['Overall'],
    'Average Human Score': [f'{human_overall_avg:.2f}'],
    'Average AI Score': [f'{ai_overall_avg:.2f}']
})

avg_scores_df = pd.concat([avg_scores_df, overall_row], ignore_index=True)

print(avg_scores_df)
```
